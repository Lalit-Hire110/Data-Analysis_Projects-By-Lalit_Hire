```python
import pandas as pd
import numpy as np

df = pd.read_csv("COVID clinical trials.csv")
df.head(5)
```

```python
# df.isnull().sum()
df.drop_duplicates()
df = df.dropna()
df.head(5)
```

## Descriptive Analysis

```python
# Check unique values in the 'Status' column
print("Unique trial statuses:")
print(df['Status'].unique())

# Count of trials per status
status_counts = df['Status'].value_counts()
print("\nTrial count by status:")
print(status_counts)

# Percent distribution of statuses
status_percent = df['Status'].value_counts(normalize=True) * 100
print("\nPercentage distribution by status:")
print(status_percent)
```

```python
#Check unique values in the 'study type' column
print(df['Study Type'].unique())
#count of trials per study type
study_type_counts = df['Study Type'].value_counts()
print("\nTrial Count by study type:")
print(study_type_counts)

#Percent distribution
study_type_percent = df['Study Type'].value_counts(normalize=True) * 100
print(study_type_percent)
```

```python
#checking unique values in the trial phase distribution
print("Unique trial Phase")
print(df['Phases'].unique())
#count per trial phase
phase_count = df['Phases'].value_counts()
print(phase_count)
#percent
phase_percent = df['Phases'].value_counts(normalize=True)*100
print(phase_percent)
```

```python
# print(df.columns.tolist())
print(df.columns)
```

```python
# Priniting Unique locations
# print(df['Locations'].unique())
# #count of trials by location\
location_counts = df['Locations'].value_counts().head(10)
# print("\nTop 10 countries by number of trials:")
# print(location_counts.head(10))
# #percent
location_per = df['Locations'].value_counts(normalize=True) * 100
# print(location_per.head(10))
top_countries_table = pd.DataFrame({
    'Country': location_counts.index, 
    'Number of Trials' : location_counts.values
})
print("Top 10 countries by number of trials:\n")
print(top_countries_table.to_string(index=False))
```

```python
from IPython.display import display
display(top_countries_table)
```

```python
#top 10 Most common Intervention
# print("Unique Interventions (sample):")
# print(df['Interventions'].dropna().unique()[:10]) #Preview first 10
#Count top interventions
intervention_counts = df['Interventions'].value_counts().head(10)
top_interventions_table = pd.DataFrame({
    'Interventions': intervention_counts.index,
    'Number of Trials': intervention_counts.values
})
display(top_interventions_table)
```

```python
# for the top 10 most common conditions
condition_counts = df['Conditions'].value_counts().head(10)
top_conditions_table = pd.DataFrame({
    "Condition": condition_counts.index,
    "Number of Trials": condition_counts.values
})
display(top_conditions_table)
```

```python
# Date Overview
# Convert start date to datetime format
df['Start Date'] = pd.to_datetime(df['Start Date'], errors = 'coerce')
#drop rowa with invalid or missing values
valid_dates = df.dropna(subset=['Start Date'])
#earliest and latest start dates
earliest_date = valid_dates['Start Date'].min()
latest_date = valid_dates['Start Date'].max()
print(earliest_date)
print(latest_date)
```

```python
# No of trials per month
# extract year-month from the Start date
valid_dates['Start_Month'] = valid_dates['Start Date'].dt.to_period('M')
# count trials per month
monthly_counts = valid_dates['Start_Month'].value_counts().sort_index()
#convert to data frame
timeline_table = pd.DataFrame({
    'Month': monthly_counts.index.astype(str),
    'Number of trials started': monthly_counts.values
})
display(timeline_table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Plotting
plt.figure(figsize=(12,6))
monthly_counts.plot(kind='line', marker = 'o')
plt.title('Monthly Trend of COVID Clinical Trial Starts')
plt.xlabel('Month')
plt.ylabel('Number of trials started')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_clean = df.dropna(subset=['Start Date', 'Locations'])
df_clean['Start_Month'] = df_clean['Start Date'].dt.to_period('M')
# Top 5 countries with most trials
top_countries = df_clean['Locations'].value_counts().head(5).index.tolist()
#filter only those top countries
df_top = df_clean[df_clean['Locations'].isin(top_countries)]
#group and count trials per month for each country
country_monthly  = df_top.groupby(['Locations','Start_Month']).size().reset_index(name = 'Trial Count')
# #group and count trials per month for each country

# # Plotting

# plt.figure(figsize=(14,6))
# sns.lineplot(data = country_monthly, x  = 'Start_Month', y = 'Trial Count', hue = 'Locations', marker = 'o')
# plt.title('Monthly Trials start by top 5 countries')
# plt.xlabel('Month')
# plt.ylabel('Number of Trials Startet')
# plt.xticks(rotation = 45)
# plt.tight_layout()
# plt.grid(True)
# plt.show()
```

```python
country_monthly['Start_Month'] = country_monthly['Start_Month'].astype(str)

# Plotting
plt.figure(figsize=(14,6))
sns.lineplot(data=country_monthly, x='Start_Month', y='Trial Count', hue='Locations', marker='o')
plt.title('Monthly Trials Started by Top 5 Countries')
plt.xlabel('Month')
plt.ylabel('Number of Trials Started')  # <-- You had a typo here: 'ylable'
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```
